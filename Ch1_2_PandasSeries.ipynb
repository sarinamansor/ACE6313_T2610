## 1. Create Pandas Series

```python
import pandas as pd
import numpy as np
```

```python
step_data = [3620, 7891, 9761, 3907, 4338, 5373]
print(f'step data = {step_data}')

step_counts = pd.Series(step_data, name='steps')
print('using pandas series')
print(step_counts)
```

```python
# Add a date range to the Series
step_counts.index = pd.date_range('20260408', periods=6)
print(step_counts)
```

##2. Select data by index value

```python
# Just like a dictionary
print(step_counts['2026-04-10'])
```

```python
# Or by index position--like an array
print(type(step_counts))
print(step_counts.iloc[3])
```

```python
# Select all of April
print(step_counts['2026-04'])
```

## 4. View or Convert Data Types

```python
# View the data type
print(step_counts.dtypes)
```

```python
# Convert to a float
step_counts = step_counts.astype(float)
```

```python
# View the data type
print(step_counts.dtypes)
```

##5. Invalid/Missing data

```python
# Create invalid data
step_counts[1:3] = np.nan
print(step_counts)
```

```python
# Now fill it in with zeros
step_counts = step_counts.fillna(0)
print(step_counts)
```
