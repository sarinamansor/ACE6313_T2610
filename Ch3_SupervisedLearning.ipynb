##1. Regresssion

```python
#Load Data
from pandas import read_csv

df = read_csv('https://raw.githubusercontent.com/sarinamansor/ML_Datasets/main/real_estate_valuation_dataset.csv')
```

```python
#split the data into features (X) and targets (y)
X = df.drop(columns=['House price of unit area'])
y = df['House price of unit area']
```

### Linear Regression

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split as split

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
print(f'R2 score: {model.score(X_test, y_test):.2f}')
```

###KNN Regressor

```python
#KNN Regressor
from sklearn.neighbors import KNeighborsRegressor

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
model = KNeighborsRegressor()
model.fit(X_train, y_train)
print(f'R2 score: {model.score(X_test, y_test):.2f}')
```

##2. Classification

```python
#Load Data
from pandas import read_csv
df = read_csv('https://raw.githubusercontent.com/sarinamansor/ML_Datasets/main/cdc_diabetes_small.csv')
```

```python
#split the data into features (X) and targets (y)
X = df.drop(columns=['Diabetes_binary'])
y = df['Diabetes_binary']
```

###Logistic Regression

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
model = LogisticRegression(max_iter=500) # Increased max_iter to help convergence
model.fit(X_train, y_train)
print(f'Accuracy: {model.score(X_test, y_test):.2%}')
```

###KNN Classifier

```python
#KNN Classification
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
model = KNeighborsClassifier()
model.fit(X_train, y_train)
print(f'Accuracy: {model.score(X_test, y_test):.2%}')
```

### Gausian Naive Bayes Classifier

```python
#Gaussian Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
model = GaussianNB()
model.fit(X_train, y_train)
print(f'Accuracy: {model.score(X_test, y_test):.2%}')
```

###SVM Classification

```python
#SVM Classification
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
model = SVC()
model.fit(X_train, y_train)
print(f'Accuracy: {model.score(X_test, y_test):.2%}')
```

##3. Decision Tree Classification

```python
#Decision Tree Classification
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
print(f'Train Accuracy: {model.score(X_train, y_train):.2%}')
print(f'Test Accuracy: {model.score(X_test, y_test):.2%}')
plot_tree(model, feature_names=df.columns, class_names=['0','1'], filled=True)
plt.tight_layout()
plt.show()
```

```python
#Decision Tree Classification with pre-pruning using max_depth
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
model = DecisionTreeClassifier(max_depth=2)
model.fit(X_train, y_train)
print(f'Train Accuracy: {model.score(X_train, y_train):.2%}')
print(f'Test Accuracy: {model.score(X_test, y_test):.2%}')
plot_tree(model, feature_names=df.columns, class_names=['0','1'], filled=True)
plt.tight_layout()
plt.show()
```

```python
#Decision Tree Classification with pre-pruning using max_leaf_nodes=6
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
model = DecisionTreeClassifier(max_leaf_nodes=6)
model.fit(X_train, y_train)
print(f'Train Accuracy: {model.score(X_train, y_train):.2%}')
print(f'Test Accuracy: {model.score(X_test, y_test):.2%}')
plot_tree(model, feature_names=df.columns, class_names=['0','1'], filled=True)
plt.tight_layout()
plt.show()
```

```python
#Get the feature importances
feature_importances = model.feature_importances_

#plot the feature importances
plt.bar(X.columns, feature_importances)
plt.xlabel('Feature')
plt.xticks(rotation=90)
plt.ylabel('Importance')
plt.title('Feature Importance')
plt.show()
```

##4. Ensemble Learning

###Random Forest Classification

```python
#Random Forest Classification
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
print(f'Accuracy: {model.score(X_test, y_test):.2%}')
```

###Gradient Boosting Classfier

```python
#Gradient Boosting Tree Classification
from sklearn.ensemble import GradientBoostingClassifier

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
model = GradientBoostingClassifier(random_state=42)
model.fit(X_train, y_train)
print(f'Accuracy: {model.score(X_test, y_test):.2%}')
```

###Voting Classifier

```python
#Voting Classifier
from sklearn.ensemble import VotingClassifier

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
clf1=LogisticRegression(max_iter=1000)
clf2=GaussianNB()
clf3=SVC()
model = VotingClassifier(estimators=[('lr', clf1), ('gnb', clf2), ('svc', clf3)], voting='hard')
model.fit(X_train, y_train)
print(f'Accuracy: {model.score(X_test, y_test):.2%}')
```

##5. Neural Network

```python
#MLP Classification using Neural Network
from sklearn.neural_network import MLPClassifier

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
model = MLPClassifier(random_state=42)
model.fit(X_train, y_train)
print(f'Accuracy: {model.score(X_test, y_test):.2%}')
```

##6. Machine Learning Pipeline

```python
#Machine Learning Pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
model = Pipeline([('scaler', StandardScaler()), ('clf', LogisticRegression())])
model.fit(X_train, y_train)
print(f'Accuracy: {model.score(X_test, y_test):.2%}')
```

```python
from os import pipe
#using GridSearchCV with pipeline
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.pipeline import Pipeline

pipe2 = Pipeline([('scaler', None), ('clf', KNeighborsClassifier())])
params={}
params['scaler']=[None, MinMaxScaler(), StandardScaler(), RobustScaler()]
params['clf']=[KNeighborsClassifier(), LogisticRegression(), SVC()]
kf = KFold(n_splits=5, shuffle=True, random_state=42)
grid = GridSearchCV(pipe2, params, cv=kf, n_jobs=-1, verbose=1)
grid.fit(X_train, y_train)
print(grid.best_params_)
best_model = grid.best_estimator_
best_model.fit(X_train, y_train)
print(f'Accuracy: {best_model.score(X_test, y_test):.2%}')
```
