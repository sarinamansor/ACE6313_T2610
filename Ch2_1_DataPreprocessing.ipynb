##5. Feature Scaling

```python
import numpy as np
from pandas import read_csv
```

```python
#read data
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
df = read_csv('https://raw.githubusercontent.com/sarinamansor/ML_Datasets/main/pima-indians-diabetes.data.csv', names=names)
array = df.values
# separate array into input and output components
X = array[:,:-1]
y = array[:,-1]
```

###Min-Max Scaling: Scale each feature into the range between 0 and 1

```python
#Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1)) #initialise the scaler to [0,1]
scaler.fit(X) #compute the min and max of X
scaledX = scaler.transform(X) #apply the min-max formula
# Check min and max of all columns
print(f'minimum={np.min(scaledX, axis=0)}, maximum={np.max(scaledX, axis=0)}')
```

```python
#plot scaledX
import matplotlib.pyplot as plt
plt.plot(scaledX)
plt.show()
```

###Standardization: Scale each feature into the range with a mean of 0 and a variance of 1

```python
#Standardization data (0 mean, 1 stdev)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
scaledX = scaler.transform(X)
# Check mean and standard deviation of all columns
print(f'mean={np.mean(scaledX, axis=0)}, variance={np.var(scaledX, axis=0)}')
```

###Robust Scaling: Scale each feature into the range with a median of 0 and interquartile range of 1

```python
#Robust scaling (0 median, 1 IQR)
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
scaler.fit(X)
scaledX = scaler.transform(X)

#check median and IQR of all columns
q3, q1 = np.percentile(scaledX, [75 ,25], axis=0)
print(f'median={np.median(scaledX, axis=0)}, IQR={q3-q1}')
```

##7. Feature Selection

###Univariate Selection

```python
#Feature Selection with Univariate Statistics
from pandas import read_csv
from sklearn.feature_selection import SelectKBest

#read data
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
df = read_csv('https://raw.githubusercontent.com/sarinamansor/ML_Datasets/main/pima-indians-diabetes.data.csv', names=names)
array = df.values
# separate array into input and output components
X = array[:,:-1]
y = array[:,-1]

selector = SelectKBest(k=4)
features = selector.fit_transform(X, y)
selected = selector.get_support()
#show selected features
print([names[i] for i in range(len(names)-1) if selected[i]])
```

###Model-Based Selection

```python

#Model-based Feature Selection with Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

selector = SelectFromModel(RandomForestClassifier(), threshold='median')
features = selector.fit_transform(X, y)
selected = selector.get_support()
#show selected features
print([names[i] for i in range(len(names)-1) if selected[i]])
```

###Recursive Featue Elimination

```python
#Feature Selection with RFE
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
rfe = RFE(model, n_features_to_select=4)
features = rfe.fit_transform(X, y)
selected = rfe.get_support()
# Show selected features
print([names[i] for i in range(len(names)-1) if selected[i]])
```
