## 1. Creating NumPy array

```python
import numpy as np
```

```python
# Create a NumPy array from a Python list
array = np.array([1, 2, 3])
print(f'array = {array}')
print(f'type = {type(array)}')

A = [1, 2, 3]
print(f'A = {A}')
print(f'type = {type(A)}')
```

```python
# Create a 5-integer array filled with zeros
a = np.zeros(5, dtype=int)
print(f'a = {a}')
```

```python
# Create a 3x2 floating-point array filled with 1s
b = np.ones((3, 2), dtype=int)
print('b=')
print(b)
```

```python
# Create a 2x3 array filled with 3.14
c = np.full((2, 3), 3.14)
print(c)
```

```python
# Create an array filled with linear sequence
# Starting at 0, ending at 3, stepping by 0.5
d = np.arange(0, 3, 0.5)
print(d)
```

```python
# Create an array of 5 values evenly spaced from 0 to 1
e = np.linspace(0, 1, 5)
print(e)
```

```python
# Create a 3x3 array of uniformly distributed random
# values between 0 and 1
f = np.random.random((3, 3))
print(f)
```

## 2. Array attributes, indexing and slicing

```python
x = np.random.randint(0, 10, (2, 3))
print(x)
print(f'dimensionality = {x.ndim}') # dimensionality
print(f'shape = {x.shape}') # shape of array
print(f'number of element = {x.size}') # number of element
```

```python
print(x[0, 1]) # element at row 0, column 1
print(x[1, -1]) # element at row 1, last column (column 2)
print(x[1, :]) # elements at row 1
print(x[:, -1]) # elements at last column (column 2)
print(x[:2, 1:]) # elements at row 0-1, column 1-2
```

## 3. Reshaping of arrays

```python
# Create a 10-integer array from 0 to 9
y = np.arange(0, 10)
print(y)
```

```python
print(y.reshape(2, 5)) # reshape to 2x5
```

```python
#reshape to an array of 5 rows
#number of columns will be automatically calculated
print(y.reshape(5,-1))
```

##4. Array joining and splitting

```python
z = np.arange(0, 6).reshape(2, 3)
print(z)
```

```python
#stack array vertically
aa = np.vstack([z, z])
print(aa)
```

```python
# Stack arrays horizontally
bb = np.hstack([z, z])
print(bb)
```

```python
# Split array vertically
upper, lower = np.vsplit(aa, [1])
print(upper)
```

```python
# Split array horizontally
left, right = np.hsplit(aa, [2])
print(right)
```
