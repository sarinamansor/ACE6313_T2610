#**Chapter 1 Part 4: Matplotlib**

##1. Plot visualisation

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-np.pi, np.pi, 50)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, 'r*-', label= 'sin')
plt.plot(x, y_cos, 'bo-', label= 'cos')
plt.xlabel('x coordinates')
plt.ylabel('y coordinates')
plt.title('A Matplotlib example')
plt.legend(loc='best')
plt.grid(axis='both')
plt.show()
```

##2. Scatter Plot

```python
pts = np.random.randint(0, 20, (10, 2))
plt.scatter(pts[:, 0], pts[:, 1], color='b', marker='*')
plt.xlabel('x-coord')
plt.ylabel('y-coord')
plt.title('A Matplotlib scatterplot example')
plt.grid(axis='both')
plt.show()
```

##3. Histogram

```python
x = ['A', 'B', 'C', 'F']
y = [21, 42, 29, 8]
plt.bar(x, y, color=['b', 'g', 'g', 'r'])
plt.xlabel('Grade')
plt.ylabel('No. of students')
plt.title('Summary of results')
plt.show()
```

##5. Sine and Cosine Waves

```python
x = np.linspace(-np.pi, np.pi, 25)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2, 1, 1)
plt.plot(x, y_sin, 'r.-')
plt.title('Sine')
plt.subplot(2, 1, 2)
plt.plot(x, y_cos, 'b^-')
plt.title('Cosine')
plt.tight_layout()
plt.show()
```
